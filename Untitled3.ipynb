```python
import regex as re
tweets = df['text'].tolist()
new = ''
new_tweets = []
# print(tweets[0])
num_pattern = '[0-9]'
user_pattern = '@\S+'
punctuation_pattern = '[^\w\s]'
hyperlink_pattern = 'http\S*'



for tweet in tweets:
  # print('------new------')
  # print(f'before ---> {tweet}')
  new = re.sub(user_pattern, '', tweet)
  
  new = re.sub(hyperlink_pattern, '', new)
 
  new = re.sub(num_pattern, '',new)
  
  new = re.sub(punctuation_pattern, '', new)
  new = new.replace("\r\n",' ')
  new = new.lower()
  # print(f'after ---> {new}')
  new_tweets.append(new)

```

```python
tweet_words = []


for tweet in new_tweets:
  
  tweet = tweet.split(' ')
  
  # print(' ')
  # print(tweet)
  tweet_words += tweet

tweet_words = list(set(tweet_words))
tweet_words.remove('')
```

```python
tweet_words = np.array(tweet_words)
```

```python
bow = []

def get_bow(tweet):
  

  tweet = tweet.split(' ')
  tweet = list(set(tweet))
  if '' in tweet:

    tweet.remove('')
  tweet = np.array(tweet)
  # print(tweet_words)
  # print(tweet)

  bag = np.zeros(len(tweet_words))

  arr, x, y = np.intersect1d(tweet_words, tweet, return_indices=True)
  bag[x] = 1
  return bag

```

```python
bow = np.array(bow)
```

```python
def calculate_distances(x_train,x_test, euclidean = True):
  len_x = x_train.shape[0]
  len_y = x_test.shape[0]

  dists = np.zeros((len_x,len_y))



  

  if euclidean:

    for i in range(len_x):
      
      a = x_train[i]
      b = np.transpose(x_test)
      # print(a.shape)
      # print(b.shape)
      c = np.square(a + np.transpose(b))
      print(c.shape)
      d = np.sqrt(np.sum(c,axis = 1))
      print(d)
      
      dists[i] = d

  else:
    for i in range(len_x):
      
      a = x_train[i]
      b = np.transpose(x_test)
      # print(a.shape)
      # print(b.shape)
      c = np.abs(a - np.transpose(b))
      d = np.sum(c,axis = 1)
      
      dists[i] = d




  # print(dists)
  return dists
```
